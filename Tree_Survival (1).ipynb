```python
# Transformando as colunas String em colunas numericas
df_t2 = pd.get_dummies(df, columns=['Subplot', 'Species', 'Light_Cat', 'Soil', 'Adult', 'Sterile',
                                   'Conspecific', 'Myco', 'SoilMyco', 'PlantDate', 'Harvest', 'Alive'])
```

```python
df_t = df_t2.drop('No', axis=1)
df_t
```

```python
# Verificando quantidade de valores Vazios em cada coluna
df.isnull().sum()
```

```python
# Preenchendo os dados vazios da coluna EMF e da coluna Event
df_t['EMF'].fillna(df_t['EMF'].mean(), inplace=True)
df_t['Event'].fillna(df_t['Event'].mean(), inplace=True)
```

```python
df_t['Event'].isnull().sum()
```

```python
# Separando os targets das features
train_x = df_t.drop(columns=['Alive_X'], axis = 1)
train_y = df_t['Alive_X']
```

```python
# Dividindo em treino e teste
train_x, test_x, train_y, test_y = train_test_split(train_x,train_y, test_size = 0.2)
```

```python
# Iniciando o treinamento
LR = LinearRegression().fit(train_x, train_y)
```

```python
# Fazendo a predição do teste
LR_predict = LR.predict(test_x)
```

```python
# Exibindo o percentual de acerto
print("R2 LR:", r2_score(test_y, LR_predict), "\n")
```
